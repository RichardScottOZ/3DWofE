```python
# Three-dimensional weights of evidence prospectivity modeling
# The input is a csv file with the following columns respectively: X, Y, Z, Grade (Target Element), Fac1, Fac2, ...
# -----------------------------------------------------------------------------------------------------------------
import csv
import math
import pandas as pd
import os
from numpy.random import rand
# seed random number generator
# generate random numbers between 0-1

# Number of voxels for different items
# NumT: total number of voxels
input_dir = 'C:/users/rscott/Downloads/'
input_file = "2792-2021-10-22-09-52-Invert-the-Gawler-Mira_Inversion_Micromine_28353.csv"
input_reader = pd.read_csv(os.path.join(input_dir,input_file) )
NumT = float(input_reader.shape[0])
print(NumT)

values = (rand(int(NumT))*2 - 1)/10
input_reader['density2'] = input_reader['density2'] + values
```

```python
input_reader.to_csv('C:/users/rscott/Downloads/woftest.csv', index=False)
```

```python
#index column property of interest = density
threshold = 2.75
NumD = float(input_reader.loc[input_reader['density'] > threshold].shape[0])
print(NumD)
```

```python
thresholds = [2.79] # Fill the list with your desired numbers
NumB = [0]*len(thresholds)

for i in range(len(thresholds)):
    NumB[i] = float(input_reader.loc[input_reader.iloc[:,i+4] > thresholds[i]].shape[0])   
    
print(NumB)    
```

```python
# NumBD: number of intersected mineralization-bearing voxels and anomalous voxels of evidential models
NumBD = [0]*len(thresholds)
for i in range(len(thresholds)):
    NumBD[i] = float(input_reader.loc[(input_reader['density'] > threshold) & (input_reader.iloc[:,i+4] > thresholds[i])].shape[0])   
print(NumBD)
```

```python
# Required probabilities, odds and logits
ProBD = []
for n in NumBD:
    ProBD.append(n/NumD)
print("ProBD",ProBD)

# "_abs" means in the absence of the last parameter before underscore
ProBD_abs = []
i = 0
for n in NumB:
    ProBD_abs.append((n-NumBD[i])/(NumT-NumD))
    i += 1
print("ProBD_abs",ProBD_abs)

ProB_absD = []
for n in NumBD:
    ProB_absD.append((NumD-n)/NumD)
print("ProB_absD",ProB_absD)    

ProB_absD_abs = []
i = 0
for n in NumB:
    ProB_absD_abs.append((NumT-n-NumD+NumBD[i])/(NumT-NumD))
    i += 1
print("ProB_absD_abs",ProB_absD_abs)    

LS = []
i = 0
for p in ProBD:
    LS.append(p/ProBD_abs[i])
    i += 1
print("LS",LS)        

W_pos = []
for i in LS:
    W_pos.append(math.log(i))
print("W_pos",W_pos)            

LN = []
i = 0
for p in ProB_absD:
    LN.append(p/ProB_absD_abs[i])
    i += 1
print("LN",LN)                

W_neg = []
for i in LN:
    W_neg.append(math.log(i))
print("W_neg",W_neg)                    

prior_p = NumD/NumT
print("prior_p",prior_p)                    

prior_o = prior_p/(1-prior_p)
print("prior_o",prior_o)                    

prior_l = math.log(prior_o)
print("prior_l",prior_l)                    

post_o_DB = []
for i in LS:
    post_o_DB.append(prior_o*i)
print("post_o_DB",post_o_DB)                    

post_l_DB = []
for i in W_pos:
    post_l_DB.append(math.log(prior_o)+i)
print("post_l_DB",post_l_DB)                        

post_o_DB_abs = []
for i in LN:
    post_o_DB_abs.append(prior_o*i)
print("post_o_DB_abs",post_o_DB_abs)                            

post_l_DB_abs = []
for i in W_neg:
    post_l_DB_abs.append(math.log(prior_o)+i)
print("post_l_DB_abs",post_l_DB_abs)                                

post_p_DB = []
for i in post_o_DB:
    post_p_DB.append(i/1+i)
print("post_p_DB",post_p_DB)                                

post_p_DB_abs = []
for i in post_o_DB_abs:
    post_p_DB_abs.append(i/1+i)
print("post_p_DB_abs",post_p_DB_abs)                                    
```
